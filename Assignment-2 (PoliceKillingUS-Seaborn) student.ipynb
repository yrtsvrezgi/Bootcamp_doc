## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn
plt.figure(figsize=(12,6))
sns.swarmplot(x = "gender", y = "age", data = kill, hue = "manner_of_death");
```

## 2: Plot 12 most dangerous cities

```python
kill["city"].value_counts().head(12)
```

```python
# with seaborn barplot
dangerous_city = kill["city"].value_counts().head(12).index
dangerous_city_no = kill["city"].value_counts().head(12).values

plt.figure(figsize=(10,5))
sns.barplot(x =dangerous_city  ,y=dangerous_city_no, data = kill, ci = None)
plt.title("Most dangerous cities", color = "b")
plt.xticks(rotation = 60);
```

```python


# with seaborn countplot
plt.figure(figsize=(10,5))
sns.countplot(x = "city", data = kill,
              order = kill["city"].value_counts().head(12).index)

plt.title("Most dangerous cities", color = "b")
plt.xticks(rotation = 60);
```

## 3: Plot 20 most dangerous states

```python
kill["state"].value_counts().head(20)
```

```python
# with seaborn barplot

dangerous_state = kill["state"].value_counts().head(20).index
dangerous_state_no = kill["state"].value_counts().head(20).values

plt.figure(figsize=(10,5))
sns.barplot(x =dangerous_state  ,y=dangerous_state_no, data = kill, ci = None)
plt.title("Most dangerous states", color = "b")
plt.xticks(rotation = 60);
```

```python
# with seaborn countplot
plt.figure(figsize=(10,5))
sns.countplot(x = "state", data = kill,
              order = kill["state"].value_counts().head(20).index)

plt.title("Most dangerous cities", color = "b")
plt.xticks(rotation = 60);
```
